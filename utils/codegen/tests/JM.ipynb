```
from funcodegen import flatten_piecewise
lut1 = Symbol(r'u_{t1}')
lut2 = Symbol(r'u_{t2}')
lun = Symbol(r'u_n')
lrt1 = Symbol(r'r_{t1}')
lrt2 = Symbol(r'r_{t2}')
lrn = Symbol(r'r_n')
epsilon = Symbol(r'\epsilon')
radius = Symbol(r'\R')
deltan = Symbol(r'\Delta_n')
deltat1 = Symbol(r'\Delta_{t1}')
deltat2 = Symbol(r'\Delta_{t2}')
ndelta = Symbol(r'\lVert {\Delta_t} \rVert')
lrhon = Symbol(r'\rho_n')
lrhot1 = Symbol(r'\rho_{t1}')
lrhot2 = Symbol(r'\rho_{t2}')
```

```
def make_readable(expr):

    return simplify(flatten_piecewise(expr).subs(rn - rhon*un, deltan).subs(rt1-rhot1*ut1, deltat1).subs(rt2-rhot2*ut2, deltat2).subs(sqrt(deltat1**2+deltat2**2),ndelta).subs(Pow(deltat1**2+deltat2**2, Rational(3, 2)), ndelta**3).subs( mu * max0D0, radius).subs(np.finfo(float).eps, epsilon).subs(rhon,lrhon).subs(rhot1, lrhot1).subs(rhot2, lrhot2).subs(rt1, lrt1).subs(rt2, lrt2).subs(rn, lrn).subs(ut1, lut1).subs(ut2, lut2).subs(un, lun))
```

```
with open('../tex/fJM.tex','w') as f:
     f.write(latex(make_readable(FAC)))
```
